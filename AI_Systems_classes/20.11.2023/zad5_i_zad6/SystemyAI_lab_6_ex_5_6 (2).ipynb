```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('Boston.csv')
print(df)
```

```python

df.head()
```

```python
rm=df.iloc[:,6]
```

```python
rm
```

```python
medv=df.iloc[:,14]
```

```python
medv
```

```python
plt.figure(figsize=(20, 5))

features = ['rm']
target = df['medv']

for i, col in enumerate(features):
    plt.subplot(1, len(features) , i+1)
    x = df[col]
    y = target
    plt.scatter(x, y, marker='o')
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('medv')
```

```python
real_rm = np.array(rm)
real_medv = np.array(medv)
```

```python
real_rm
```

```python
import tensorflow as tf
import keras
from keras.layers import Dense
from keras.models import Sequential
```

```python
model=Sequential()
```

```python
model.add(Dense(units = 1, use_bias=True, input_dim=1, activation = "linear"))
```

```python
opt = tf.keras.optimizers.Adam(learning_rate=0.1)
```

```python
model.compile(loss='MSE',optimizer=opt)
```

```python
model.summary()
```

```python
epochs = 1500
h = model.fit(real_rm,real_medv, verbose=1, epochs=epochs, batch_size=150, validation_split=0.3)
```

```python
#Loss = h.history['loss']
#Loss
```

```python
plt.scatter(np.arange(epochs),h.history['loss'])
plt.scatter(np.arange(epochs),h.history['val_loss'],c='r')
plt.show()
```

```python
weights = model.get_weights()
print(weights[0][0][0])
print(weights[1][0])

```

```python
max = np.max(rm)
min = np.min(rm)
X = np.linspace(min, max, num=10)
plt.plot(X,weights[0][0][0]*X+weights[1][0],c='r')
plt.scatter(rm,medv,c="b")
plt.show()
```
