```
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

```
import keras
from keras.models import Sequential
from keras.layers import Dense
```

Zbi√≥r danych:

```
x_label1 = np.random.normal(3, 1, 1000)
y_label1 = np.random.normal(2, 1, 1000)
x_label2 = np.random.normal(7, 1, 1000)
y_label2 = np.random.normal(6, 1, 1000)

xs = np.append(x_label1, x_label2)
ys = np.append(y_label1, y_label2)
labels = np.asarray([0.]*len(x_label1)+[1.]*len(x_label2))
```

```
plt.scatter(x_label1, y_label1, c='r', marker='x', s=20)
plt.scatter(x_label2, y_label2, c='g', marker='1', s=20)
plt.show()
```

Funkcja b≈Çƒôdu (entropia krzy≈ºowa):

```
def loss_fn(label, label_model):
    return tf.reduce_mean(-label*tf.math.log(label_model)-(1-label)*tf.math.log(1-label_model))
```

PoczƒÖtkowe warto≈õci parametr√≥w i pƒôtla uczƒÖca:

```
import random
a = tf.Variable(random.random())
b = tf.Variable(random.random())

c = tf.Variable(random.random())

Loss = []
epochs = 3000
lr = 0.1

for _ in range(epochs):

  with tf.GradientTape() as tape:
    #predykcja modelu

    pred_y = tf.sigmoid(a * xs + b * ys + c)
    #policzenie b≈Çƒôdu
    loss = loss_fn(labels, pred_y)
    #zapisanie aktualnej warto≈õci b≈Çƒôdu
    Loss.append(loss.numpy())

  dloss_da, dloss_db, dloss_dc = tape.gradient(loss,(a, b,c))

  a.assign_sub(lr*dloss_da)  #a = a - alpha*dloss_da
  b.assign_sub(lr*dloss_db)  #b = b - alpha*dloss_db
  c.assign_sub(lr*dloss_dc)
```

```
Loss
```

```
plt.plot(Loss)
plt.show()
```

Na uczenie modelu ma najwiekszy wp≈Çyw u≈ºycie batcha (bez batcha jest podawany ca≈Çy zbi√≥r uczƒÖcy), dziƒôki temu wprowadza, ≈ºe pewnƒÖ losowo≈õƒá w procesie uczenia, pomoga to uniknƒÖƒá utkniƒôcia w minimach lokalnych. Model uczony z minibatchem osiƒÖga lepsze rezultaty je≈ºeli chodzi o wyniki uczenia(szybszy spadek funkcji b≈Çƒôdu oraz mniejszy b≈ÇƒÖd). Model lepiej i szybciej siƒô uczy gdy mini-batch jest wiƒôkszy ni≈º gdy jest on mniejszy.

Ponadto na proces uczenia modelu ma wp≈Çyw ilo≈õƒá epok.
Za ma≈Ça ilo≈õƒá epok skutkuje niedouczeniem modelu (model nie nauczy≈Ç siƒô wystarczajƒÖco dobrze dostosowywaƒá siƒô do danych treningowych).
Ostatnim sprawdzonym przeze mnie parametrem jest wsp√≥≈Çczynnik uczenia. Po przestawieniu na wsp√≥≈Çczynnik Adam model uczy siƒô lepiej
Jego zbyt du≈ºa warto≈õƒá rowadzi do skakania wok√≥≈Ç minimum globalnego przy czym model go nie osiƒÖgnie. W przypadku zastosowania zbyt ma≈Çej warto≈õci wsp√≥≈Çczynnika uczenia proces uczenia jest bardzo wolny na prze≈Çomie epok, a model "utyka" w minimach lokalnych.

```
Loss = []
Val_loss =[]
epochs = 1000
learning_rate = 0.01
batch_size = 50
val_split = 0.3
a = tf.Variable(random.random())
b = tf.Variable(random.random())
c = tf.Variable(random.random())
x_tr,y_tr,label_tr,x_val,y_val,label_val = val_train_split(xs,ys,labels,val_split)
for _ in range(epochs):
  xs_batch,ys_batch,labels_batch = subset_dataset_2(x_tr,y_tr,label_tr,batch_size)
  xsv_batch,ysv_batch,labelsv_batch = subset_dataset_2(x_val,y_val,label_val,batch_size)
  with tf.GradientTape() as tape:
    pred_l = tf.sigmoid(a * xs_batch + b * ys_batch + c)
    pred_lv = tf.sigmoid(a * xsv_batch + b * ysv_batch + c)
    #print(label_batch.shape)
    loss = loss_fn(labels_batch, pred_l)
    val_loss = loss_fn(labelsv_batch,pred_lv)
    Loss.append(loss.numpy())
    Val_loss.append(val_loss.numpy())

  dloss_da, dloss_db, dloss_dc = tape.gradient(loss,(a, b,c))

  a.assign_sub(learning_rate*dloss_da)  #a = a - alpha*dloss_da
  b.assign_sub(learning_rate*dloss_db)  #b = b - alpha*dloss_db
  c.assign_sub(learning_rate*dloss_dc)
```
